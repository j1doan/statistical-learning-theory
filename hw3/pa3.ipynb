## Part A: Mini Course on Concentration Inequalities

### 1. The concept of concentration.
Concentration inequalities control fluctuations of a random quantity around its mean (or median).  
Typical form: if $Z$ depends on many random inputs, then
$$
\Pr\big(|Z-\mathbb E Z|\ge t\big)\le \text{small}(t).
$$
The phrase *concentrates* means this tail probability decays quickly (often exponentially in $t^2$ or $t$).

For learning, the central object is the empirical risk
$$
L_n(h)=\frac1n\sum_{i=1}^n\ell(h;Z_i),\qquad Z_i\stackrel{i.i.d.}{\sim}\mathcal D,\ \ell\in[0,1].
$$
Since $L_n(h)$ is an average of bounded random variables, we expect it to be close to
$$
L_\mathcal D(h)=\mathbb E[\ell(h;Z)]
$$
with high probability. Concentration inequalities make this precise.

**Concrete example (coin flips).** If $X_i\in\{0,1\}$ are Bernoulli($p$), then $\bar X_n=\frac1n\sum X_i$ estimates $p$. Hoeffding$^{1}$ gives
$$
\Pr\big(|\bar X_n-p|\ge \varepsilon\big)\le 2e^{-2n\varepsilon^2}.
$$
So error $\varepsilon$ has probability roughly $e^{-n\varepsilon^2}$: doubling sample size dramatically shrinks tail risk.

**Intuition.** Independent noise terms cancel. To make an average deviate by $t$, many terms must align in the same unfavorable direction, an exponentially unlikely event.

### 2. The common technique.
The unifying mechanism is the **MGF/Chernoff method**.

Let $Z$ be any random variable and $\lambda>0$. By Markov,
$$
\Pr(Z\ge t)=\Pr(e^{\lambda Z}\ge e^{\lambda t})\le e^{-\lambda t}\,\mathbb E[e^{\lambda Z}].
$$
Hence
$$
\Pr(Z\ge t)\le \inf_{\lambda>0}\exp\!\big(-\lambda t+\log \mathbb E e^{\lambda Z}\big).
$$
So a tail bound is obtained once we control the log-MGF $\psi_Z(\lambda):=\log\mathbb E e^{\lambda Z}$.

For sums $Z=\sum_i Y_i$ of independent variables, MGFs factor:
$$
\psi_Z(\lambda)=\sum_i \log\mathbb E[e^{\lambda Y_i}],
$$
which reduces the proof to one-variable bounds.

- Hoeffding: uses boundedness $Y_i\in[a_i,b_i]$ to show sub-Gaussian log-MGF.  
- Bernstein: uses boundedness **and** variance information to get a refined log-MGF and sharper small-deviation behavior.

For functions $f(X_1,\dots,X_n)$ (not necessarily sums), the martingale variant applies:
1. Define Doob martingale $M_i=\mathbb E[f\mid X_1,\dots,X_i]$.
2. Bound increments $|M_i-M_{i-1}|$ by bounded differences constants.
3. Apply exponential supermartingale/Chernoff to $M_n-M_0$.

This yields McDiarmid's inequality.

### 3. Statements, assumptions, and consequences.
Let $X_1,\dots,X_n$ be random variables and $S_n=\sum_{i=1}^n(X_i-\mathbb E X_i)$.

| Inequality | Assumptions | Controls | Typical tail |
|---|---|---|---|
| Hoeffding | independent, $X_i\in[a_i,b_i]$ a.s. | sum/mean of bounded independent RVs | $\Pr(S_n\ge t)\le \exp\!\left(-\frac{2t^2}{\sum_i(b_i-a_i)^2}\right)$ |
| Hoeffding (without replacement) | sample uniformly without replacement from finite population in $[a,b]$ | sample mean from finite population | at least as concentrated as i.i.d. sampling; e.g. $\Pr(\bar X-\mu\ge\varepsilon)\le e^{-2n\varepsilon^2/(b-a)^2}$ (and sharper Serfling finite-population correction) |
| McDiarmid | independent inputs, bounded differences: changing coordinate $i$ changes $f$ by at most $c_i$ | general function $f(X_1,\dots,X_n)$ | $\Pr(f-\mathbb Ef\ge t)\le \exp\!\left(-\frac{2t^2}{\sum_i c_i^2}\right)$ |
| Bernstein | independent, centered, $|X_i|\le b$, variance proxy $V=\sum\mathbb E[X_i^2]$ | sum of bounded RVs with variance info | $\Pr(S_n\ge t)\le \exp\!\left(-\frac{t^2}{2(V+bt/3)}\right)$ |

**Comparison.**
- Hoeffding uses only support bounds, so it is distribution-agnostic but can be conservative.
- Bernstein adds variance: when variance is small, it improves the rate (sub-Gaussian for moderate $t$, sub-exponential only in the far tail).
- McDiarmid extends concentration from sums to *stable functions* of independent inputs.
- Sampling without replacement gains negative dependence; concentration is no worse (often better) than i.i.d. sampling.

### 4. Proofs.
#### 4.1 Hoeffding (independent bounded case)
If $Y\in[m,M]$ and $\mathbb EY=0$, Hoeffding's lemma gives
$$
\log\mathbb E e^{\lambda Y}\le \frac{\lambda^2(M-m)^2}{8}.
$$
For $Y_i=X_i-\mathbb E X_i$ (independent),
$$
\log\mathbb E e^{\lambda S_n}\le \sum_i \frac{\lambda^2(b_i-a_i)^2}{8}.
$$
Chernoff then gives
$$
\Pr(S_n\ge t)\le \exp\!\left(-\lambda t+\frac{\lambda^2}{8}\sum_i(b_i-a_i)^2\right).
$$
Optimizing in $\lambda$ yields
$$
\Pr(S_n\ge t)\le \exp\!\left(-\frac{2t^2}{\sum_i(b_i-a_i)^2}\right).\ \square
$$

#### 4.2 Hoeffding for sampling without replacement
Let $(X_1,\dots,X_n)$ be sampled without replacement from a finite multiset $\{v_1,\dots,v_N\}\subset[a,b]$, and let $(Y_1,\dots,Y_n)$ be i.i.d. draws with replacement from the same population. Hoeffding (1963) proves convex domination
$$
\mathbb E\,\phi\!\left(\sum_{i=1}^n X_i\right)\le \mathbb E\,\phi\!\left(\sum_{i=1}^n Y_i\right)
$$
for every convex $\phi$. Taking $\phi(u)=e^{\lambda u}$ gives MGF domination; therefore every Chernoff tail for i.i.d. bounded sampling also holds without replacement. Applying Hoeffding's i.i.d. bound to $Y_i$ yields
$$
\Pr(\bar X-\mu\ge\varepsilon)\le \exp\!\left(-\frac{2n\varepsilon^2}{(b-a)^2}\right).\ \square
$$
(Serfling's refinement multiplies exponent by a finite-population correction factor.)

#### 4.3 McDiarmid (bounded differences)
Assume independent $X_1,\dots,X_n$ and
$$
\sup_{x_1,\dots,x_n,x_i'}|f(x_1,\dots,x_i,\dots,x_n)-f(x_1,\dots,x_i',\dots,x_n)|\le c_i.
$$
Define $M_i=\mathbb E[f(X_1,\dots,X_n)\mid X_1,\dots,X_i]$. Then $(M_i)$ is a martingale and $|M_i-M_{i-1}|\le c_i$. Applying Hoeffding's lemma conditionally to each increment gives
$$
\log\mathbb E e^{\lambda(M_n-M_0)}\le \frac{\lambda^2}{8}\sum_i c_i^2.
$$
Chernoff optimization yields
$$
\Pr\big(f-\mathbb Ef\ge t\big)\le \exp\!\left(-\frac{2t^2}{\sum_i c_i^2}\right).\ \square
$$

#### 4.4 Bernstein
Let independent centered $X_i$ satisfy $|X_i|\le b$ a.s., and set $V=\sum_i\mathbb E[X_i^2]$. A standard mgf bound is
$$
\log\mathbb E e^{\lambda X_i}\le \frac{\lambda^2\mathbb E[X_i^2]}{2(1-\lambda b/3)},\qquad 0<\lambda<3/b.
$$
Summing over $i$:
$$
\log\mathbb E e^{\lambda S_n}\le \frac{\lambda^2V}{2(1-\lambda b/3)}.
$$
Then
$$
\Pr(S_n\ge t)\le \inf_{0<\lambda<3/b}\exp\!\left(-\lambda t+\frac{\lambda^2V}{2(1-\lambda b/3)}\right)
$$
and optimization gives
$$
\Pr(S_n\ge t)\le \exp\!\left(-\frac{t^2}{2(V+bt/3)}\right).\ \square
$$

**Common template across all four proofs:** exponential tilt + MGF control + optimization in $\lambda$. The only difference is *how* the MGF is controlled (bounded range, variance refinement, martingale increments, or negative dependence).

### 5. Connection to the new ERM proof.

We establish the connection to the new ERM proof via one of the following methods *(Note: Method 1 is more traditional, Method 2 may be more intuitive depending on the reader)*:

**Method 1: Using Hoeffding's Inequality and Bernstein’s Inequality**

Let losses be in $[0,1]$. Define
$$
h^*\in\arg\min_{h\in\mathcal H}L_\mathcal D(h),\qquad g_h(z):=\ell(h;z)-\ell(h^*;z)\in[-1,1].
$$
Then $\Delta(h):=\mathbb E[g_h]=L_\mathcal D(h)-L_\mathcal D(h^*)\ge 0$, and for ERM $\hat h\in\arg\min_h L_n(h)$ we have
$$
\Delta(\hat h)=L_\mathcal D(\hat h)-L_\mathcal D(h^*).
$$

For $0$-$1$ loss (and similarly for bounded losses with a constant factor), the excess variable satisfies a Bernstein condition
$$
\operatorname{Var}(g_h)\le \mathbb E[g_h]=\Delta(h).
$$
Applying Bernstein to fixed $h$ and then a two-sample symmetrization/union bound over labelings on $2n$ points gives, with probability at least $1-\delta$, simultaneously for all $h\in\mathcal H$,
$$
\Delta(h)\le 2\big(L_n(h)-L_n(h^*)\big)+C\frac{\log \Gamma_\mathcal H(2n)+\log(1/\delta)}{n}
$$
for a universal constant $C>0$.

Now plug in $h=\hat h$. Since $\hat h$ minimizes $L_n$,
$$
L_n(\hat h)-L_n(h^*)\le 0.
$$
Therefore
$$
L_\mathcal D(\hat h)-L_\mathcal D(h^*)=\Delta(\hat h)\le C\frac{\log \Gamma_\mathcal H(2n)+\log(1/\delta)}{n}.
$$
Equivalently,
$$
L_\mathcal D(\hat h)\le \inf_{h\in\mathcal H}L_\mathcal D(h)+O\!\left(\frac{\log \Gamma_\mathcal H(2n)+\log(1/\delta)}{n}\right).
$$
This is the announced i.i.d. growth-function ERM guarantee.

**Method 2: Using Ghost Samples**

Let losses be in $[0,1]$. Define
$$
h^*\in\arg\min_{h\in\mathcal H}L_\mathcal D(h),\qquad g_h(z):=\ell(h;z)-\ell(h^*;z)\in[-1,1].
$$
Then $\Delta(h):=\mathbb E[g_h]=L_\mathcal D(h)-L_\mathcal D(h^*)\ge 0$, and for ERM $\hat h\in\arg\min_h L_n(h)$ we have
$$
\Delta(\hat h)=L_\mathcal D(\hat h)-L_\mathcal D(h^*).
$$

To control $\Delta(h)$ uniformly we follow the standard symmetrization + counting pipeline.

**Symmetrization.** [7] Draw an independent ghost sample $S'=(Z_{n+1},\dots,Z_{2n})$ from $\mathcal D^n$ and let $L_n'(h)$ be the empirical risk on $S'$. For every class of $[0,1]$ losses (hence for $g_h$) we have
$$
\Pr\Big(\sup_{h\in\mathcal H}\big(L_\mathcal D(h)-L_n(h)\big) > \varepsilon\Big) \le 2\,\Pr\Big(\sup_{h\in\mathcal H}\big(L_n'(h)-L_n(h)\big) > \varepsilon/2\Big).
$$
Condition on $(S,S')$. For fixed $h$, $L_n'(h)-L_n(h)$ is an average of $2n$ centered, bounded variables, so Bernstein (using $\operatorname{Var}(g_h)\le \Delta(h) \le 1$) gives
$$
\Pr\Big(L_n'(h)-L_n(h) > t \mid S,S'\Big) \le \exp\!\left(-c n t^2\right)
$$
for some universal $c>0$.

**Counting via $\Gamma_\mathcal H(2n)$.** The gap $L_n'(h)-L_n(h)$ depends only on the restriction of $h$ to the $2n$ points in $S\cup S'$. Hence, conditioned on $(S,S')$, there are at most $\Gamma_\mathcal H(2n)$ distinct values of the gap. A union bound gives
$$
\Pr\Big(\sup_{h\in\mathcal H}|L_n'(h)-L_n(h)| > t \mid S,S'\Big) \le \Gamma_\mathcal H(2n)\exp\!\left(-c n t^2\right).
$$
Integrating over $(S,S')$ and combining with symmetrization yields (with probability $\ge 1-\delta$) the uniform bound
$$
\sup_{h\in\mathcal H}|L_\mathcal D(h)-L_n(h)| \le \sqrt{\frac{1}{c n}\Big(\log \Gamma_\mathcal H(2n)+\log\frac{2}{\delta}\Big)}.
$$

**Plugging in ERM optimality.** On the above $1-\delta$ event,
$$
L_\mathcal D(\hat h) \le L_n(\hat h)+\sqrt{\frac{1}{c n}\Big(\log \Gamma_\mathcal H(2n)+\log\frac{2}{\delta}\Big)} \le L_n(h^*)+\sqrt{\frac{1}{c n}\Big(\log \Gamma_\mathcal H(2n)+\log\frac{2}{\delta}\Big)}.
$$
A second application of the uniform deviation bound to $h^*$ replaces $L_n(h^*)$ by $L_\mathcal D(h^*)$, resulting in
$$
L_\mathcal D(\hat h) \le L_\mathcal D(h^*)+2\sqrt{\frac{1}{c n}\Big(\log \Gamma_\mathcal H(2n)+\log\frac{2}{\delta}\Big)}.
$$
Absorbing constants into $O(\cdot)$ gives precisely the stated growth-function ERM guarantee.

## Part B

### 1. Proof.
We prove the updated No-Free-Lunch theorem.

**Quantifiers.** The learner $A$ and integer $n<|\mathcal X|/2$ are fixed universally. The adversary then chooses a hard distribution $\mathcal D$.

Choose distinct points $C=\{x_1,\dots,x_{2n}\}\subset\mathcal X$. For each labeling $\sigma\in\{0,1\}^{2n}$ define $f_\sigma(x_i)=\sigma_i$. Let $\mathcal D_\sigma$ be uniform on
$$
\{(x_i,f_\sigma(x_i)):1\le i\le 2n\}.
$$
Clearly $L_{\mathcal D_\sigma}(f_\sigma)=0$ (realizable).

Let $S\sim \mathcal D_\sigma^n$. For fixed sample points, define unseen set
$$
U(S):=\{i\in[2n]:x_i\ \text{does not appear in }S\}.
$$
Conditioned on sample points, labels on $U(S)$ are independent fair bits under uniform random $\sigma$, while $A(S)$ is independent of those unseen bits. Hence each unseen point contributes error probability $1/2$. Therefore
$$
\mathbb E_\sigma\big[L_{\mathcal D_\sigma}(A(S))\mid S\big]=\frac{|U(S)|}{4n}.
$$
Taking expectation over $S$:
$$
\mathbb E_{\sigma,S}\,L_{\mathcal D_\sigma}(A(S))=\frac{\mathbb E|U(S)|}{4n}.
$$
Now
$$
\mathbb E|U(S)|=\sum_{i=1}^{2n}\Pr(x_i\text{ unseen})=2n\left(1-\frac1{2n}\right)^n\ge 2n e^{-1/2}.
$$
So
$$
\mu:=\mathbb E_{\sigma,S}\,L_{\mathcal D_\sigma}(A(S))\ge \frac{e^{-1/2}}{2}>0.3.
$$
By averaging, there exists at least one labeling $\sigma^*$ such that
$$
\mathbb E_S\,L_{\mathcal D_{\sigma^*}}(A(S))\ge \mu.
$$
Set $Y:=L_{\mathcal D_{\sigma^*}}(A(S))\in[0,1]$. Using
$$
\mathbb E[Y]\le a\,\Pr(Y<a)+1\cdot\Pr(Y\ge a)=a+(1-a)\Pr(Y\ge a),
$$
we get
$$
\Pr(Y\ge a)\ge \frac{\mathbb E[Y]-a}{1-a}.
$$
With $a=1/8$ and $\mathbb E[Y]\ge 0.3$,
$$
\Pr\!\left(Y\ge \frac18\right)\ge \frac{0.3-0.125}{0.875}=0.2\ge \frac17.
$$
Thus for $\mathcal D=\mathcal D_{\sigma^*}$: (i) realizable, and (ii) with probability at least $1/7$, $L_\mathcal D(A(S))\ge 1/8$. $\ \square$

### 2. Application: closing the Fundemental Theorem.
**Corollary (NFL $\Rightarrow$ lower-bound direction).** If $\operatorname{VCdim}(\mathcal H)=\infty$, then $\mathcal H$ is not PAC learnable (distribution-free, realizable binary classification).

**Reason.** Fix any learner $A$ and sample size $n$. Since $\operatorname{VCdim}(\mathcal H)=\infty$, pick a shattered set of size $2n$. Restricting $\mathcal H$ to that set realizes all $2^{2n}$ labelings, so the NFL theorem applies and gives a realizable distribution with
$$
\Pr\big(L_\mathcal D(A(S))\ge 1/8\big)\ge 1/7.
$$
Hence no finite $n$ can guarantee arbitrary $(\varepsilon,\delta)$ (e.g., $\varepsilon<1/8$, $\delta<1/7$), so PAC learnability fails.

**Concrete infinite-VC class.** Let $\mathcal X=\mathbb N$ and
$$
\mathcal H=\{0,1\}^{\mathbb N}\quad\text{(all binary functions on }\mathbb N\text{)}.
$$
For any finite set $S=\{x_1,\dots,x_m\}\subset\mathbb N$ and any labels $y\in\{0,1\}^m$, define $h_y\in\mathcal H$ by $h_y(x_i)=y_i$ and arbitrary values elsewhere. So every finite $S$ is shattered, therefore $\operatorname{VCdim}(\mathcal H)=\infty$. Applying the corollary gives that this class is not PAC learnable.

### 3. Worked construction.
Take $n=1$, domain $\mathcal X=\{a,b,c\}$ (so $n<|\mathcal X|/2$), and learning rule:
- On sample $((x,y))$, output hypothesis $h_{x,y}$ with $h_{x,y}(x)=y$ and $h_{x,y}(x')=0$ for $x'\neq x$.

Choose target $f^*(a)=f^*(b)=1$ (and any value on $c$), and distribution $\mathcal D$ uniform on $\{(a,1),(b,1)\}$. Then $L_\mathcal D(f^*)=0$ (realizable).

If sample point is $a$, learner predicts $h(a)=1,h(b)=0$; if sample point is $b$, learner predicts $h(a)=0,h(b)=1$. In either case, exactly one of $a,b$ is misclassified, each with mass $1/2$. Therefore
$$
L_\mathcal D(A(S))=\frac12\quad\text{for every sample }S.
$$
Hence
$$
\Pr_{S\sim\mathcal D}\!\left(L_\mathcal D(A(S))\ge\frac18\right)=1.
$$
This explicitly exhibits a bad distribution/target pair for this learner, with constant (indeed unit) failure probability.

### 4. Comparison to the original NFL theorem.
| Aspect | Original theorem (mistake bound) | Updated theorem (PAC/statistical) |
|---|---|---|
| Setting | Online/sequential prediction on a finite domain | Batch statistical learning under i.i.d. sampling |
| Adversary | Chooses an ordering/sequence and labels to force mistakes (can be adaptive in online framing) | Chooses a fixed distribution $\mathcal D$ (oblivious) before sampling |
| Learner type | Deterministic learner in sequence model | Any learner (possibly randomized) in PAC model |
| Conclusion type | Deterministic worst-case number of mistakes on a sequence | Probabilistic lower bound on population loss over random sample |
| Quantity lower-bounded | Total mistakes (discrete count) | Risk $L_\mathcal D(A(S))$ (expected 0-1 error under data distribution) |

**Interpretation.**
- Original NFL: without structural assumptions, there exists a sequence where the learner performs catastrophically in online prediction.
- Updated NFL: without finite-capacity assumptions (finite VC dimension), there exists a realizable distribution on which learning cannot uniformly guarantee small generalization error.

Both are called "No-Free-Lunch" because each says: absent inductive bias/complexity control, there is no universally successful learner.
