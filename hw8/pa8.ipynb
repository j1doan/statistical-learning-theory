## Part A: Choosing Regularization by Validation


### 1. Validation selection for a finite random candidate set.

**Claim:** $\mathbb{E}_V L_{\mathcal{D}}(h_{\hat{k}}) \le \min_{k \in [K]} L_{\mathcal{D}}(h_k) + \sqrt{\frac{\log(2K)}{2n_V}}$

**Proof:**

Let $L_{\mathcal{D}}(h)$ be the true risk and $L_V(h)$ be the empirical risk on validation set $V$.
Since the predictors $h_k$ are fixed and independent of $V$, for each $k \in [K]$, the expected value of $L_V(h_k)$ over $V$ is exactly the true risk $L_{\mathcal{D}}(h_k)$.

Let us define the deviation for each candidate as $Z_k = L_{\mathcal{D}}(h_k) - L_V(h_k)$. Since the loss function is bounded, $\ell(w, z) \in [0, 1]$, both the true risk and the empirical risk are bounded in $[0, 1]$. Therefore, the difference $Z_k$ has zero mean and, by Hoeffding's lemma for bounded variables, is sub-Gaussian with variance proxy $1 / (4n_V)$. This means for any $s > 0$:
$$ \mathbb{E}_V[\exp(s Z_k)] \le \exp\left(\frac{s^2}{8 n_V}\right) $$

We are interested in bounding the expected maximum deviation, so let $Z = \max_{k \in [K]} Z_k$. Using Jensen's inequality (since the exponential function is convex) and the fact that the maximum of non-negative elements is bounded by their sum, we have:
$$ \exp(s \mathbb{E}_V[Z]) \le \mathbb{E}_V[\exp(s Z)] = \mathbb{E}_V\left[\max_{k \in [K]} \exp(s Z_k)\right] \le \sum_{k=1}^K \mathbb{E}_V[\exp(s Z_k)] \le K \exp\left(\frac{s^2}{8 n_V}\right) $$

Taking the natural logarithm of both sides and dividing by $s > 0$:
$$ \mathbb{E}_V[Z] \le \frac{\log K}{s} + \frac{s}{8 n_V} $$

To minimize this bound, we plug in $s = \sqrt{8 n_V \log K}$, which yields:
$$ \mathbb{E}_V[Z] \le \sqrt{\frac{\log K}{2 n_V}} $$

Now, let $k^* \in \arg\min_{k \in [K]} L_{\mathcal{D}}(h_k)$ be the index of the predictor with the minimum true risk. For the empirically selected predictor $\hat{k}$, we can bound its true risk:
$$ L_{\mathcal{D}}(h_{\hat{k}}) = L_V(h_{\hat{k}}) + (L_{\mathcal{D}}(h_{\hat{k}}) - L_V(h_{\hat{k}})) \le L_V(h_{\hat{k}}) + Z $$

Because $\hat{k}$ was chosen specifically to minimize the empirical validation risk $L_V$, we know that $L_V(h_{\hat{k}}) \le L_V(h_{k^*})$. Thus:
$$ L_V(h_{\hat{k}}) + Z \le L_V(h_{k^*}) + Z $$

Taking the expectation over the random validation set $V$:
$$ \mathbb{E}_V [L_{\mathcal{D}}(h_{\hat{k}})] \le \mathbb{E}_V [L_V(h_{k^*})] + \mathbb{E}_V [Z] $$

Since $h_{k^*}$ is fixed independently of $V$, its expected empirical risk is its true risk. Substituting our bound for $\mathbb{E}_V[Z]$:
$$ \mathbb{E}_V [L_{\mathcal{D}}(h_{\hat{k}})] \le L_{\mathcal{D}}(h_{k^*}) + \sqrt{\frac{\log K}{2 n_V}} = \min_{k \in [K]} L_{\mathcal{D}}(h_k) + \sqrt{\frac{\log K}{2 n_V}} $$

Since $K \le 2K$, this implies the slightly looser bound given in the claim:
$$ \mathbb{E}_V [L_{\mathcal{D}}(h_{\hat{k}})] \le \min_{k \in [K]} L_{\mathcal{D}}(h_k) + \sqrt{\frac{\log (2K)}{2 n_V}} $$


### 2. Oracle inequality over a grid.

**Claim:** $\mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le \min_{\lambda \in \Lambda} \inf_{u \in \mathcal{W}} \left\{ L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} \right\} + \sqrt{\frac{\log(2K)}{2n_V}}$

**Proof:**

From the setup, the sample is split into independent training and validation sets, $T$ and $V$. Conditionally on the training sample $T$, the trained rules $h_\lambda$ for $\lambda \in \Lambda$ are fixed and entirely independent of $V$. Therefore, we can apply the result from Part 1 directly with $K = |\Lambda|$:
$$ \mathbb{E}_V[L_{\mathcal{D}}(h_{\hat{\lambda}}) \mid T] \le \min_{\lambda \in \Lambda} L_{\mathcal{D}}(h_\lambda) + \sqrt{\frac{\log(2K)}{2n_V}} $$

Now, taking the expectation over the random training sample $T$:
$$ \mathbb{E}[L_{\mathcal{D}}(h_{\hat{\lambda}})] = \mathbb{E}_T [\mathbb{E}_V[L_{\mathcal{D}}(h_{\hat{\lambda}}) \mid T]] \le \mathbb{E}_T \left[ \min_{\lambda \in \Lambda} L_{\mathcal{D}}(h_\lambda) \right] + \sqrt{\frac{\log(2K)}{2n_V}} $$

By Jensen's inequality (or simply exchanging the expectation and the minimum, since the minimum of expectations is $\ge$ the expectation of the minimum), we have:
$$ \mathbb{E}_T \left[ \min_{\lambda \in \Lambda} L_{\mathcal{D}}(h_\lambda) \right] \le \min_{\lambda \in \Lambda} \mathbb{E}_T [L_{\mathcal{D}}(h_\lambda)] $$

For any fixed $\lambda \in \Lambda$, $h_\lambda$ is the Regularized Empirical Risk Minimizer (RERM) trained on $T$. By the standard Learning from Stable RERM theorem, since $\ell$ is $G$-Lipschitz and $\Psi$ is $\alpha$-strongly convex, for any arbitrary comparator $u \in \mathcal{W}$:
$$ \mathbb{E}_T [L_{\mathcal{D}}(h_\lambda)] \le L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} $$

Because this holds for any $u \in \mathcal{W}$, we can take the infimum over $u$:
$$ \mathbb{E}_T [L_{\mathcal{D}}(h_\lambda)] \le \inf_{u \in \mathcal{W}} \left\{ L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} \right\} $$

Substituting this back into our expectation bound and taking the minimum over $\lambda \in \Lambda$:
$$ \min_{\lambda \in \Lambda} \mathbb{E}_T [L_{\mathcal{D}}(h_\lambda)] \le \min_{\lambda \in \Lambda} \inf_{u \in \mathcal{W}} \left\{ L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} \right\} $$

Combining all steps yields the desired oracle inequality:
$$ \mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le \min_{\lambda \in \Lambda} \inf_{u \in \mathcal{W}} \left\{ L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} \right\} + \sqrt{\frac{\log(2K)}{2n_V}} $$

**Explanation:**
Selecting among $K$ trained RERM rules means we evaluate each rule on the validation set to pick the best one. However, the empirical validation loss is only a random estimate of the true risk. The union bound over all $K$ candidates forces us to pay a complexity penalty of $\sqrt{\frac{\log K}{n_V}}$ (or $\sqrt{\frac{\log(2K)}{2n_V}}$ specifically here). This term acts as a uniform convergence bound, ensuring that our empirical estimate of the validation loss faithfully tracks the true risk across all candidate rules, preventing us from "overfitting" the validation set.


### 3. Adapting to an unknown comparator scale.

**Claim:** $\mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le L_{\mathcal{D}}(u) + \frac{5\sqrt{2}}{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n_T}} + \sqrt{\frac{\log(2K)}{2n_V}}$

**Proof:**

Let $u \in \mathcal{W}$ be a comparator such that $B_{\min}^2 \le \Psi(u) \le B_{\max}^2$. Let $a = \Psi(u) > 0$. The two-term expression we want to optimize is:
$$ f(\lambda) = \lambda a + \frac{2G^2}{\lambda\alpha n_T} $$

If we could choose a continuous value of $\lambda$, taking the derivative with respect to $\lambda$ and setting it to zero gives the optimal continuous choice:
$$ \lambda_{\text{opt}}(a) = \sqrt{\frac{2G^2}{\alpha a n_T}} $$

At this optimal value, the two terms are perfectly balanced:
$$ \lambda_{\text{opt}}(a) a = \sqrt{\frac{2G^2 a}{\alpha n_T}} \quad \text{and} \quad \frac{2G^2}{\lambda_{\text{opt}}(a)\alpha n_T} = \sqrt{\frac{2G^2 a}{\alpha n_T}} $$
So the minimum continuous value is $2\sqrt{\frac{2G^2 a}{\alpha n_T}}$.

However, we are restricted to a finite grid $\Lambda$. The problem states that the grid has the property that there exists some $\lambda \in \Lambda$ satisfying:
$$ \frac{1}{2}\lambda_{\text{opt}}(a) \le \lambda \le 2\lambda_{\text{opt}}(a) $$
We can represent this chosen grid point as $\lambda = c \lambda_{\text{opt}}(a)$ for some constant $c \in [1/2, 2]$.

Substituting $\lambda = c \lambda_{\text{opt}}(a)$ into our two-term expression:
$$ \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} = c \lambda_{\text{opt}}(a) a + \frac{2G^2}{c \lambda_{\text{opt}}(a)\alpha n_T} = c \sqrt{\frac{2G^2 a}{\alpha n_T}} + \frac{1}{c} \sqrt{\frac{2G^2 a}{\alpha n_T}} = \left(c + \frac{1}{c}\right) \sqrt{\frac{2G^2 a}{\alpha n_T}} $$

We need to find the worst-case penalty for this discretization over the interval $c \in [1/2, 2]$. The function $g(c) = c + \frac{1}{c}$ is strictly convex for $c > 0$. Therefore, its maximum on any closed interval must occur at one of the boundaries. Checking the boundaries:
- For $c = 1/2$, $g(1/2) = 1/2 + 2 = 2.5 = 5/2$.
- For $c = 2$, $g(2) = 2 + 1/2 = 2.5 = 5/2$.

Thus, for our chosen grid point $\lambda$, the expression is bounded by:
$$ \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n_T} \le \frac{5}{2} \sqrt{\frac{2G^2 a}{\alpha n_T}} = \frac{5}{2} \sqrt{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n_T}} = \frac{5\sqrt{2}}{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n_T}} $$

Using the oracle inequality from Part 2, which states that the expected loss is bounded by the minimum over all $\lambda \in \Lambda$, we can safely upper bound this minimum by evaluating it at the specific $\lambda$ we just analyzed. This yields the final bound:
$$ \mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le L_{\mathcal{D}}(u) + \frac{5\sqrt{2}}{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n_T}} + \sqrt{\frac{\log(2K)}{2n_V}} $$


### 4. A dyadic grid and the price of tuning.

**Claim:** A dyadic grid of size $K = \lceil \log_2(B_{\max}/B_{\min}) \rceil + 1$ satisfies the condition, leading to the simplified guarantee:
$$ \mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le L_{\mathcal{D}}(u) + 5 \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n}} + \sqrt{\frac{\log(2K)}{n}} $$

**Proof:**

The required grid must cover the full range of optimal $\lambda$ values for any comparator norm bounded by $a \in [B_{\min}^2, B_{\max}^2]$. Since $\lambda_{\text{opt}}(a)$ is inversely proportional to $\sqrt{a}$, the extreme values of $\lambda$ we need to cover are:
$$ \lambda_{\max} = \lambda_{\text{opt}}(B_{\min}^2) = \frac{\sqrt{2}G}{B_{\min}\sqrt{\alpha n_T}}, \quad \lambda_{\min} = \lambda_{\text{opt}}(B_{\max}^2) = \frac{\sqrt{2}G}{B_{\max}\sqrt{\alpha n_T}} $$

A simple way to satisfy the requirement that a grid point $\lambda$ exists within a factor of 2 of $\lambda_{\text{opt}}(a)$ is to use a dyadic grid (powers of 2). We can construct the grid starting from $\lambda_{\max}$ and halving it repeatedly:
$$ \Lambda = \{\lambda_{\max} 2^{-k} : k = 0, 1, \dots, K-1\} $$

To ensure that even the smallest possible optimal value $\lambda_{\min}$ has a grid point $\lambda$ satisfying $\lambda \le 2\lambda_{\min}$, the smallest point in our grid must be at most $2\lambda_{\min}$:
$$ \lambda_{\max} 2^{-(K-1)} \le 2\lambda_{\min} \implies \lambda_{\max} 2^{-K} \le \lambda_{\min} \implies 2^{K} \ge \frac{\lambda_{\max}}{\lambda_{\min}} $$

Substituting our definitions for $\lambda_{\max}$ and $\lambda_{\min}$:
$$ 2^{K} \ge \frac{\sqrt{2}G / (B_{\min}\sqrt{\alpha n_T})}{\sqrt{2}G / (B_{\max}\sqrt{\alpha n_T})} = \frac{B_{\max}}{B_{\min}} $$

Thus, taking the base-2 logarithm, the number of grid points required is exactly $K = \lceil \log_2(B_{\max}/B_{\min}) \rceil + 1$ (which is $O(\log(B_{\max}/B_{\min}))$).

Now, assuming $n$ is even and setting the data split to be $n_T = n_V = \frac{n}{2}$, we substitute this into our bound from Part 3:
$$ \frac{5\sqrt{2}}{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha (n/2)}} = \frac{5\sqrt{2}}{2} \frac{\sqrt{2}G\sqrt{\Psi(u)}}{\sqrt{\alpha n}} = \frac{10}{2} \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n}} = 5 \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n}} $$
And for the validation term:
$$ \sqrt{\frac{\log(2K)}{2(n/2)}} = \sqrt{\frac{\log(2K)}{n}} $$

This gives the simplified guarantee:
$$ \mathbb{E} L_{\mathcal{D}}(h_{\hat{\lambda}}) \le L_{\mathcal{D}}(u) + 5 \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n}} + \sqrt{\frac{\log(2K)}{n}} $$

**Identification of terms:**
* **RERM learning term:** $5 \frac{G\sqrt{\Psi(u)}}{\sqrt{\alpha n}}$. This term represents the core statistical error of learning with an appropriately tuned regularized empirical risk minimizer. Notice that it is penalized (inflated by a constant factor) because we are forced to use only half the data for training and because our grid discretization is slightly suboptimal.
* **Validation-selection term:** $\sqrt{\frac{\log(2K)}{n}}$. This bounds the error from evaluating $K$ different candidate rules on the finite validation set. It is the cost of searching for the best parameter.
* **Additional term caused by not knowing comparator scale:** Because we do not know the correct scale $\Psi(u)$ in advance, we cannot simply fix an optimal $\lambda$. We are forced to split our data (halving the effective $n$) and tune over a grid. This directly introduces the entire validation-selection penalty $\sqrt{\frac{\log(2K)}{n}}$, and simultaneously inflates the RERM learning term from its ideal $2\sqrt{2}$ factor to $5$.


## Part B: Approximate RERM and Optimization Error


### 1. Approximate minimizers are close to exact minimizers.

**Claim:** $\|\tilde{w}_S - w_S\| \le \sqrt{\frac{2\eta}{\lambda\alpha}}$

**Proof:**

Consider the regularized objective function $F_S(w) = L_S(w) + \lambda\Psi(w)$. This function is the sum of a convex empirical loss $L_S(w)$ and a regularizer $\lambda\Psi(w)$. Because $\Psi$ is assumed to be $\alpha$-strongly convex with respect to the norm $\|\cdot\|$, scaling it by $\lambda > 0$ means that $\lambda\Psi(w)$ is $\lambda\alpha$-strongly convex. Since the sum of a convex function and a strongly convex function is strongly convex, $F_S(w)$ itself is $\lambda\alpha$-strongly convex.

Let $w_S \in \arg\min_{w \in \mathcal{W}} F_S(w)$ be the exact minimizer over the convex domain $\mathcal{W}$. By the definition of strong convexity, for any arbitrary $w \in \mathcal{W}$, the function satisfies:
$$ F_S(w) \ge F_S(w_S) + \langle \nabla F_S(w_S), w - w_S \rangle + \frac{\lambda\alpha}{2} \|w - w_S\|^2 $$

Because $w_S$ is the minimizer over the convex set $\mathcal{W}$, the first-order optimality condition dictates that moving in any feasible direction from $w_S$ cannot decrease the objective. Therefore, the inner product of the gradient and any feasible direction is non-negative: $\langle \nabla F_S(w_S), w - w_S \rangle \ge 0$. Substituting this gives a simpler lower bound:
$$ F_S(w) \ge F_S(w_S) + \frac{\lambda\alpha}{2} \|w - w_S\|^2 $$

Now, we apply this property to the approximate minimizer $\tilde{w}_S$ returned by our algorithm:
$$ F_S(\tilde{w}_S) - F_S(w_S) \ge \frac{\lambda\alpha}{2} \|\tilde{w}_S - w_S\|^2 $$

We are provided with a deterministic optimization guarantee that the algorithm's objective value is at most $\eta$ worse than the optimal: $F_S(\tilde{w}_S) - F_S(w_S) \le \eta$. Combining these inequalities allows us to directly bound the parameter distance:
$$ \frac{\lambda\alpha}{2} \|\tilde{w}_S - w_S\|^2 \le \eta \implies \|\tilde{w}_S - w_S\|^2 \le \frac{2\eta}{\lambda\alpha} \implies \|\tilde{w}_S - w_S\| \le \sqrt{\frac{2\eta}{\lambda\alpha}} $$


### 2. Stability of approximate RERM.

**Claim:** For $S$ and $S'$ differing in one example, $|\ell(\tilde{w}_S, z) - \ell(\tilde{w}_{S'}, z)| \le \frac{2G^2}{\lambda\alpha n} + 2G\sqrt{\frac{2\eta}{\lambda\alpha}}$

**Proof:**

We want to bound the difference in loss for our approximate RERM when trained on two samples $S$ and $S'$ that differ by exactly one point. First, we will bound the parameter distance between the two approximate minimizers, $\tilde{w}_S$ and $\tilde{w}_{S'}$, using the triangle inequality for the parameter norm:
$$ \|\tilde{w}_S - \tilde{w}_{S'}\| = \|\tilde{w}_S - w_S + w_S - w_{S'} + w_{S'} - \tilde{w}_{S'}\| $$
$$ \|\tilde{w}_S - \tilde{w}_{S'}\| \le \|\tilde{w}_S - w_S\| + \|w_S - w_{S'}\| + \|w_{S'} - \tilde{w}_{S'}\| $$

We have bounds for each of these three terms:
1.  From Part 1, the distance from the approximate minimizer on $S$ to the exact minimizer on $S$ is $\|\tilde{w}_S - w_S\| \le \sqrt{\frac{2\eta}{\lambda\alpha}}$.
2.  Similarly, on sample $S'$, the algorithm guarantees $\|\tilde{w}_{S'} - w_{S'}\| \le \sqrt{\frac{2\eta}{\lambda\alpha}}$.
3.  We are given the exact RERM parameter movement bound for uniform replacement stability: because $S$ and $S'$ differ in one example, the exact minimizers shift by at most $\|w_S - w_{S'}\| \le \frac{2G}{\lambda\alpha n}$.

Summing these bounds gives the total parameter movement for the approximate RERM:
$$ \|\tilde{w}_S - \tilde{w}_{S'}\| \le 2\sqrt{\frac{2\eta}{\lambda\alpha}} + \frac{2G}{\lambda\alpha n} $$

The loss function $\ell(w, z)$ is assumed to be $G$-Lipschitz with respect to the norm $\|\cdot\|$. This Lipschitz property allows us to translate the parameter distance directly into a bound on the loss difference for any test point $z$:
$$ |\ell(\tilde{w}_S, z) - \ell(\tilde{w}_{S'}, z)| \le G \|\tilde{w}_S - \tilde{w}_{S'}\| $$

Substituting our established bound for the parameter distance:
$$ |\ell(\tilde{w}_S, z) - \ell(\tilde{w}_{S'}, z)| \le G \left( \frac{2G}{\lambda\alpha n} + 2\sqrt{\frac{2\eta}{\lambda\alpha}} \right) = \frac{2G^2}{\lambda\alpha n} + 2G\sqrt{\frac{2\eta}{\lambda\alpha}} $$


### 3. Learning guarantee with optimization error.

**Claim:** $\mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n} + G\sqrt{\frac{2\eta}{\lambda\alpha}}$

**Proof:**

To establish a risk bound for the approximate minimizer, we relate its true risk back to the true risk of the exact minimizer. For any single test point $z$, using the $G$-Lipschitz property of the loss function $\ell$, the loss of the approximate minimizer is bounded by the loss of the exact minimizer plus a penalty for the distance between them:
$$ \ell(\tilde{w}_S, z) - \ell(w_S, z) \le |\ell(\tilde{w}_S, z) - \ell(w_S, z)| \le G \|\tilde{w}_S - w_S\| $$
$$ \ell(\tilde{w}_S, z) \le \ell(w_S, z) + G \|\tilde{w}_S - w_S\| $$

The true risk $L_{\mathcal{D}}(w)$ is simply the expectation of the loss over $z \sim \mathcal{D}$. Taking the expectation of both sides over $z$ gives:
$$ L_{\mathcal{D}}(\tilde{w}_S) \le L_{\mathcal{D}}(w_S) + G \|\tilde{w}_S - w_S\| $$

Next, we want to find the expected true risk by taking the expectation over the random draw of the training sample $S \sim \mathcal{D}^n$. By linearity of expectation:
$$ \mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le \mathbb{E}_S L_{\mathcal{D}}(w_S) + G \, \mathbb{E}_S \|\tilde{w}_S - w_S\| $$

From Part 1, we proved a deterministic bound that holds for every sample $S$: $\|\tilde{w}_S - w_S\| \le \sqrt{\frac{2\eta}{\lambda\alpha}}$. We can substitute this worst-case bound into the expectation:
$$ \mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le \mathbb{E}_S L_{\mathcal{D}}(w_S) + G\sqrt{\frac{2\eta}{\lambda\alpha}} $$

Finally, we recognize that $\mathbb{E}_S L_{\mathcal{D}}(w_S)$ is exactly the expected risk of the exact RERM rule. According to the Learning from Stable RERM theorem, for any comparator $u \in \mathcal{W}$, this exact RERM risk is bounded by:
$$ \mathbb{E}_S L_{\mathcal{D}}(w_S) \le L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n} $$

Substituting this RERM bound into our inequality gives the final expected true-risk bound for the approximate minimizer:
$$ \mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n} + G\sqrt{\frac{2\eta}{\lambda\alpha}} $$


### 4. How accurate must the optimizer be?

**Claim:** A sufficient condition on $\eta$ is $\eta \le \frac{G B}{\sqrt{2 \alpha} n^{3/2}}$, which yields an excess-risk bound of $L_{\mathcal{D}}(u) + \frac{(2\sqrt{2} + 1)GB}{\sqrt{\alpha n}}$.

**Proof:**

We are given a specific optimal scaling for the regularization parameter: $\lambda = \sqrt{\frac{2G^2}{\alpha B^2 n}} = \frac{\sqrt{2}G}{B\sqrt{\alpha n}}$. We want to compete with all comparators $u$ satisfying $\Psi(u) \le B^2$.

First, let's simplify the first three terms of our bound from Part 3 under these conditions:
$$ L_{\mathcal{D}}(u) + \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n} $$
Since $\Psi(u) \le B^2$, we can upper bound $\lambda\Psi(u)$ by $\lambda B^2$. Substituting our specific choice of $\lambda$:
$$ \lambda B^2 + \frac{2G^2}{\lambda\alpha n} = \left(\frac{\sqrt{2}G}{B\sqrt{\alpha n}}\right) B^2 + \frac{2G^2}{\left(\frac{\sqrt{2}G}{B\sqrt{\alpha n}}\right)\alpha n} $$
$$ = \frac{\sqrt{2}GB}{\sqrt{\alpha n}} + \frac{2G^2 \cdot B\sqrt{\alpha n}}{\sqrt{2}G \alpha n} = \frac{\sqrt{2}GB}{\sqrt{\alpha n}} + \frac{\sqrt{2}GB}{\sqrt{\alpha n}} = \frac{2\sqrt{2}GB}{\sqrt{\alpha n}} $$
So the base exact RERM bound is $L_{\mathcal{D}}(u) + \frac{2\sqrt{2}GB}{\sqrt{\alpha n}}$.

Now, we are asked to find a sufficient condition on the optimization accuracy $\eta$ such that the optimization error term is bounded by $\frac{GB}{\sqrt{\alpha n}}$:
$$ G\sqrt{\frac{2\eta}{\lambda\alpha}} \le \frac{GB}{\sqrt{\alpha n}} $$
Dividing by $G$ and squaring both sides:
$$ \frac{2\eta}{\lambda\alpha} \le \frac{B^2}{\alpha n} $$
Multiplying by $\lambda\alpha / 2$:
$$ \eta \le \frac{\lambda B^2}{2 n} $$

To make this condition explicit in terms of fundamental parameters, we substitute the definition of $\lambda$ back in:
$$ \eta \le \left(\frac{\sqrt{2}G}{B\sqrt{\alpha n}}\right) \frac{B^2}{2 n} = \frac{\sqrt{2} G B^2}{2 B n \sqrt{\alpha n}} = \frac{G B}{\sqrt{2 \alpha} n^{3/2}} $$
This is our sufficient condition for $\eta$. If the optimizer is this accurate, the penalty introduced by approximation is well-controlled.

Under this condition, the total excess-risk bound simply adds the $\frac{GB}{\sqrt{\alpha n}}$ penalty to the simplified exact RERM bound:
$$ \mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le L_{\mathcal{D}}(u) + \frac{2\sqrt{2}GB}{\sqrt{\alpha n}} + \frac{GB}{\sqrt{\alpha n}} = L_{\mathcal{D}}(u) + \frac{(2\sqrt{2} + 1)GB}{\sqrt{\alpha n}} $$


### 5. Soft-margin linear classification.

**Claim:** The approximate-RERM bound is $L_{\mathcal{D}}(u) + \frac{2RB}{\sqrt{n}} + \sqrt{R B \eta \sqrt{n}}$, and a sufficient condition on $\eta$ is $\eta \le \frac{R B}{n^{3/2}}$.

**Proof:**

For soft-margin SVM, the loss function is the hinge loss: $\ell(w, (x, y)) = (1 - y\langle w, x \rangle)_+$. For feature vectors bounded by $\|x\|_2 \le R$, the hinge loss is known to be $G$-Lipschitz with $G = R$ in the Euclidean norm $\|\cdot\|_2$.

The regularizer is Euclidean weight decay: $\Psi(w) = \frac{1}{2}\|w\|_2^2$. The function $\frac{1}{2}\|w\|_2^2$ is exactly 1-strongly convex with respect to $\|\cdot\|_2$, so the strong convexity parameter is $\alpha = 1$.

We want to compete with a comparator $u$ bounded by $\|u\|_2 \le B$. This implies a bound on the regularizer: $\Psi(u) = \frac{1}{2}\|u\|_2^2 \le \frac{1}{2}B^2$. In the context of Part 4, our new "B-squared" term is actually $\tilde{B}^2 = \frac{B^2}{2}$.

Let's specialize the optimal $\lambda$ from Part 4 using these specific values ($G = R$, $\alpha = 1$, and target scale $\frac{B^2}{2}$):
$$ \lambda = \sqrt{\frac{2R^2}{1 \cdot (\frac{B^2}{2}) \cdot n}} = \sqrt{\frac{4R^2}{B^2 n}} = \frac{2R}{B\sqrt{n}} $$

Now, substituting these into the three exact RERM terms from Part 3:
$$ \lambda\Psi(u) + \frac{2G^2}{\lambda\alpha n} \le \left(\frac{2R}{B\sqrt{n}}\right) \left(\frac{B^2}{2}\right) + \frac{2R^2}{\left(\frac{2R}{B\sqrt{n}}\right) \cdot 1 \cdot n} = \frac{RB}{\sqrt{n}} + \frac{2R^2 B \sqrt{n}}{2 R n} = \frac{RB}{\sqrt{n}} + \frac{RB}{\sqrt{n}} = \frac{2RB}{\sqrt{n}} $$

Next, let's specialize the optimization-error term $G\sqrt{\frac{2\eta}{\lambda\alpha}}$:
$$ G\sqrt{\frac{2\eta}{\lambda\alpha}} = R\sqrt{\frac{2\eta}{\left(\frac{2R}{B\sqrt{n}}\right) \cdot 1}} = R\sqrt{\frac{2\eta B\sqrt{n}}{2R}} = R\sqrt{\frac{\eta B\sqrt{n}}{R}} = \sqrt{R B \eta \sqrt{n}} $$

Combining these gives the specialized approximate-RERM bound in terms of $R, B, n,$ and $\eta$:
$$ \mathbb{E}_S L_{\mathcal{D}}(\tilde{w}_S) \le L_{\mathcal{D}}(u) + \frac{2RB}{\sqrt{n}} + \sqrt{R B \eta \sqrt{n}} $$

Finally, we find the sufficient condition on $\eta$ to ensure this optimization-error term is no larger than $\frac{RB}{\sqrt{n}}$. Setting up the inequality:
$$ \sqrt{R B \eta \sqrt{n}} \le \frac{RB}{\sqrt{n}} $$
Squaring both sides:
$$ R B \eta \sqrt{n} \le \frac{R^2 B^2}{n} $$
Solving for $\eta$:
$$ \eta \le \frac{R^2 B^2}{R B n \sqrt{n}} = \frac{R B}{n^{3/2}} $$
Thus, $\eta \le \frac{R B}{n^{3/2}}$ is the sufficient condition for soft-margin linear classification to maintain optimal statistical rates despite optimization errors.


## Part C: A Weighted Euclidean Geometry


### 1. Geometry quantities.

**Claim:** $\|\varphi(x)\|_{Q,*}^2 \le \sum_{j=1}^d \frac{r_j^2}{q_j}$ and $\sup_{w \in \mathcal{C}_b} \Psi_Q(w) = \frac{1}{2} \sum_{j=1}^d q_j b_j^2$.

**Proof:**

Let's tackle the dual norm bound first. The dual norm of the weighted Euclidean norm $\|v\|_Q = \sqrt{v^\top Q v}$ is defined algebraically as $\|v\|_{Q,*} = \sqrt{v^\top Q^{-1} v}$. 

Since our matrix $Q$ is explicitly given as a diagonal matrix $Q = \text{diag}(q_1, \dots, q_d)$ with strictly positive diagonal entries $q_j > 0$, its inverse $Q^{-1}$ is simply the diagonal matrix of the reciprocals: $Q^{-1} = \text{diag}(1/q_1, \dots, 1/q_d)$. 

Therefore, for the feature vector $\varphi(x)$, the squared dual norm expands as a sum of scaled squared components:
$$ \|\varphi(x)\|_{Q,*}^2 = \varphi(x)^\top Q^{-1} \varphi(x) = \sum_{j=1}^d \frac{\varphi_j(x)^2}{q_j} $$

We are given the component-wise feature bounds $|\varphi_j(x)| \le r_j$ for all $x$ and $j$. Squaring this inequality gives $\varphi_j(x)^2 \le r_j^2$. Since $q_j > 0$, dividing by $q_j$ preserves the inequality, allowing us to upper bound the sum:
$$ \|\varphi(x)\|_{Q,*}^2 \le \sum_{j=1}^d \frac{r_j^2}{q_j} $$

Now, let's find the supremum of the regularizer. The regularizer is defined as $\Psi_Q(w) = \frac{1}{2} w^\top Q w$. Since $Q$ is diagonal, this is just a weighted sum of squares:
$$ \Psi_Q(w) = \frac{1}{2} \sum_{j=1}^d q_j w_j^2 $$

We want to find the maximum of this function over the constrained domain $\mathcal{C}_b = \{w \in \mathbb{R}^d : |w_j| \le b_j \text{ for all } j\}$. The constraints $|w_j| \le b_j$ imply $w_j^2 \le b_j^2$. Because all weights $q_j$ are strictly positive, the sum is maximized when each individual term $w_j^2$ takes its maximum possible value, which is $b_j^2$. 

Substituting $w_j^2 = b_j^2$ yields the supremum:
$$ \sup_{w \in \mathcal{C}_b} \Psi_Q(w) = \frac{1}{2} \sum_{j=1}^d q_j b_j^2 $$


### 2. Fixed $Q$.

**Claim:** $\mathbb{E} L_{\mathcal{D}}(A_\lambda(S)) \le \inf_{w \in \mathcal{C}_b} L_{\mathcal{D}}(w) + \frac{2g}{\sqrt{n}}\sqrt{\left(\sum_{j=1}^d q_j b_j^2\right)\left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right)}$

**Proof:**

We are operating with a linear predictor where the scalar loss is $g$-Lipschitz. We first need to establish the Lipschitz constant $G$ of the loss function $\ell(w, z)$ with respect to our chosen parameter norm $\|\cdot\|_Q$. 

Using Hölder's inequality (which states $|\langle u, v \rangle| \le \|u\| \|v\|_*$), we can bound the difference in loss for two parameter vectors $w$ and $w'$:
$$ |\ell(w, z) - \ell(w', z)| \le g|\langle w - w', \varphi(x) \rangle| \le g\|w - w'\|_Q \|\varphi(x)\|_{Q,*} $$

The true Lipschitz constant $G$ with respect to $\|\cdot\|_Q$ is therefore $G = g \sup_x \|\varphi(x)\|_{Q,*}$. Using our result from Part 1, we can upper bound $G^2$:
$$ G^2 \le g^2 \sum_{j=1}^d \frac{r_j^2}{q_j} $$

The regularizer $\Psi_Q(w) = \frac{1}{2}w^\top Q w$ is perfectly tailored to be 1-strongly convex with respect to the norm $\|w\|_Q = \sqrt{w^\top Q w}$, which means our strong convexity parameter is $\alpha = 1$.

Now, we apply the Week 8 Learning from Stable RERM theorem. For any comparator $w \in \mathcal{C}_b$, the expected risk of the algorithm $A_\lambda(S)$ is bounded by:
$$ \mathbb{E} L_{\mathcal{D}}(A_\lambda(S)) \le L_{\mathcal{D}}(w) + \lambda\Psi_Q(w) + \frac{2G^2}{\lambda\alpha n} $$

To make this bound independent of a specific $w$, we can bound the regularizer term using the supremum found in Part 1, $\Psi_Q(w) \le \frac{1}{2}\sum_j q_j b_j^2$. Substituting this and our bound for $G^2$ (with $\alpha = 1$) gives:
$$ \mathbb{E} L_{\mathcal{D}}(A_\lambda(S)) \le L_{\mathcal{D}}(w) + \frac{\lambda}{2} \left(\sum_{j=1}^d q_j b_j^2\right) + \frac{2g^2}{\lambda n} \left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right) $$
Since this holds for any $w \in \mathcal{C}_b$, we can take the infimum over $w \in \mathcal{C}_b$.

Finally, we optimize this bound over our choice of the regularization strength $\lambda > 0$. The expression has the form $A\lambda + B/\lambda$, which is minimized when the two terms are equal ($A\lambda = B/\lambda \implies \lambda = \sqrt{B/A}$). The minimum value is $2\sqrt{AB}$. 
In our case, $A = \frac{1}{2}\sum_{j=1}^d q_j b_j^2$ and $B = \frac{2g^2}{n} \sum_{j=1}^d \frac{r_j^2}{q_j}$. The minimized excess term is:
$$ 2\sqrt{ \left(\frac{1}{2}\sum_{j=1}^d q_j b_j^2\right) \left(\frac{2g^2}{n} \sum_{j=1}^d \frac{r_j^2}{q_j}\right) } = 2\sqrt{ \frac{g^2}{n} \left(\sum_{j=1}^d q_j b_j^2\right) \left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right) } = \frac{2g}{\sqrt{n}}\sqrt{\left(\sum_{j=1}^d q_j b_j^2\right)\left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right)} $$

Substituting this optimized excess term back into the inequality gives the final result:
$$ \mathbb{E} L_{\mathcal{D}}(A_\lambda(S)) \le \inf_{w \in \mathcal{C}_b} L_{\mathcal{D}}(w) + \frac{2g}{\sqrt{n}}\sqrt{\left(\sum_{j=1}^d q_j b_j^2\right)\left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right)} $$


### 3. Best diagonal geometry.

**Claim:** The best excess term is $\frac{2g}{\sqrt{n}} \sum_{j=1}^d b_j r_j$.

**Proof:**

From Part 2, the excess risk term depends on the specific diagonal geometry $Q$ through the product inside the square root:
$$ \text{Excess Risk} \propto \sqrt{\left(\sum_{j=1}^d q_j b_j^2\right)\left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right)} $$
Our goal is to choose the diagonal elements $q_j > 0$ to minimize this product, thereby obtaining the tightest possible bound across all diagonal geometries.

We can apply the Cauchy-Schwarz inequality, which states that for any real sequences $u_j$ and $v_j$, $(\sum u_j^2)(\sum v_j^2) \ge (\sum u_j v_j)^2$. Let us define our sequences strategically to match our product terms:
Let $u_j = \sqrt{q_j} b_j$ and let $v_j = \frac{r_j}{\sqrt{q_j}}$. 
Note that $u_j^2 = q_j b_j^2$ and $v_j^2 = \frac{r_j^2}{q_j}$.

Applying Cauchy-Schwarz:
$$ \left(\sum_{j=1}^d u_j^2\right) \left(\sum_{j=1}^d v_j^2\right) \ge \left( \sum_{j=1}^d u_j v_j \right)^2 $$
$$ \left(\sum_{j=1}^d q_j b_j^2\right) \left(\sum_{j=1}^d \frac{r_j^2}{q_j}\right) \ge \left( \sum_{j=1}^d \left(\sqrt{q_j} b_j\right) \left(\frac{r_j}{\sqrt{q_j}}\right) \right)^2 = \left( \sum_{j=1}^d b_j r_j \right)^2 $$

This establishes a fundamental lower bound on the product that is entirely independent of our choice of $Q$. 

Furthermore, Cauchy-Schwarz tells us that this lower bound is strictly achievable (equality holds) if and only if the sequences are proportional, i.e., $u_j \propto v_j$ for all $j$. This condition means:
$$ \sqrt{q_j} b_j \propto \frac{r_j}{\sqrt{q_j}} $$
Multiplying both sides by $\sqrt{q_j}$ gives:
$$ q_j b_j \propto r_j \implies q_j \propto \frac{r_j}{b_j} $$
Since we need $q_j > 0$, setting $q_j = \frac{r_j}{b_j}$ (or any positive scalar multiple) is a valid choice that perfectly achieves the lower bound.

Substituting this achievable minimum value back into the original excess risk expression:
$$ \text{Best Excess Risk} = \frac{2g}{\sqrt{n}} \sqrt{ \left( \sum_{j=1}^d b_j r_j \right)^2 } = \frac{2g}{\sqrt{n}} \sum_{j=1}^d b_j r_j $$
This represents the absolute best-case bound achievable by tuning the diagonal geometry $Q$ to match the scales of the parameter domain $b$ and feature space $r$.
