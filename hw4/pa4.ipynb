## Part A: Sparse Linear Predictors and Model Selection

### 1. VC Dimension Through Supports

**Statement.** For the class $\mathcal{H}_k$ of $k$-sparse homogeneous halfspaces in $\mathbb{R}^d$, the VC dimension satisfies:
$$\operatorname{VCdim}(\mathcal{H}_k) = O\big(k \log(\tfrac{e d}{k})\big)$$

**Proof.** Count possible supports, apply the VC growth function bound to each $k$-dimensional subclass, and union-bound the total growth function.

1. **Decomposition by Support:** Let $I \subseteq [d]$ with $|I| = k$ be a specific support. The subclass $\mathcal{H}_I$ (halfspaces restricted to coordinates in $I$) is isomorphic to homogeneous halfspaces in $\mathbb{R}^k$. The VC dimension of each subclass is $k$.
2. **Subclass Growth Function:** By Sauer's Lemma, the growth function for any subclass $\mathcal{H}_I$ is bounded by: 
    $$\Pi_{\mathcal{H}_I}(n) \le \left(\frac{en}{k}\right)^k$$
3. **Union Bound on Growth Functions:** The class $\mathcal{H}_k$ is the union of $\binom{d}{k}$ such subclasses. The growth function of a union is bounded by the sum of the growth functions of its components:
    $$\Pi_{\mathcal{H}_k}(n) \le \sum_{I:|I|=k} \Pi_{\mathcal{H}_I}(n)$$
    Substituting the bound from Step 2, this leads to the inequality:
    $$2^n \le \binom{d}{k} \left(\frac{en}{k}\right)^k$$
4. **Solving for $n$:** To find the maximum $n$ that can be shattered, we take the logarithm of both sides and use the inequality $\log \binom{d}{k} \le k \log(\frac{ed}{k})$:
    $$n \le \log_2 \binom{d}{k} + k \log_2 \left(\frac{en}{k}\right) \le k \log_2 \left(\frac{ed}{k}\right) + k \log_2 \left(\frac{en}{k}\right)$$
5. **Conclusion:** For large $n$, the linear term $n$ on the left is constrained by the $k \log(d/k)$ term on the right. Solving this transcendental inequality yields $n = O(k \log(\frac{ed}{k}))$. Thus:
    $$\operatorname{VCdim}(\mathcal{H}_k) = O\big(k \log(\tfrac{e d}{k})\big) \quad \blacksquare$$

### 2. Penalty-based SRM

**Statement.**

With a prior $p_k$ over sparsity levels $k=1,\dots,d$ (where $\sum p_k \le 1$), define the SRM estimator $\hat{h}$ as:
$$\hat{h} \in \arg\min_{k, h \in \mathcal{H}_k} \Big\{ L_S(h) + C \sqrt{\tfrac{k \log(\tfrac{e d}{k}) + \log(\tfrac{1}{p_k}) + \log(\tfrac{1}{\delta})}{n}} \Big\}$$

With probability $\ge 1-\delta$, the generalization error satisfies:
$$L_{\mathcal{D}}(\hat{h}) \le \inf_{1 \le k \le d,\, h \in \mathcal{H}_k} \Big\{ L_{\mathcal{D}}(h) + 2C \sqrt{\tfrac{k \log(\tfrac{e d}{k}) + \log(\tfrac{1}{p_k}) + \log(\tfrac{1}{\delta})}{n}} \Big\}$$

**Proof.** Instantiate the class-level SRM theorem using the VC dimensions for sparse halfspaces.

1. **Class-Level SRM:** General SRM theory guarantees that for a collection of classes $\{\mathcal{H}_k\}$, an oracle inequality holds if the penalty accounts for the complexity of each $\mathcal{H}_k$ and the prior weight $p_k$ assigned to it.
2. **Substituting Complexity:** From Part 1, we know the VC dimension of $k$-sparse halfspaces in $d$ dimensions is $\text{VCdim}(\mathcal{H}_k) \le C_1 k \log(\frac{e d}{k})$.
3. **Decomposing the Penalty:** The term inside the square root represents a triple union bound:
    * **Support-selection cost:** $k \log(\frac{e d}{k})$ pays for searching over $\binom{d}{k}$ possible coordinate sets.
    * **Sparsity-level cost:** $\log(1/p_k)$ pays for the choice of $k$ itself.
    * **Confidence cost:** $\log(1/\delta)$ ensures the bound holds with high probability.
4. **Conclusion:** Combining these terms and absorbing universal constants into $C$ yields the stated bound. $\blacksquare$

**Statistical Cost Paid for Choosing Support vs. Sparsity Level**

The SRM framework reveals a hierarchy of "prices" that must be paid to ensure generalization when searching for sparse patterns:

* **The Support vs. The Level:** There is a fundamental distinction between the cost of finding the *right features* and the cost of choosing the *number of features*. The $k \log(d/k)$ term dominates when $d$ is large, as it represents the massive combinatorial search space of supports.
* **The Power of the Prior:** By choosing a non-uniform prior (e.g., $p_k \propto 1/k^2$), we can assign more "budget" to simpler models. This prevents a uniform penalty of $\log d$ from over-penalizing very sparse solutions, allowing the algorithm to remain sensitive to simple structures even in extremely high dimensions.
* **Occam’s Razor by Design:** The SRM objective functions as a mathematical realization of Occam's Razor. It will only accept a more complex, denser model (higher $k$) if the improvement in empirical error $L_S(h)$ is large enough to compensate for the increased statistical "cost" of the more complex support.

**Takeaway.** SRM efficiently manages the trade-off between fitting the data and the double-complexity of selecting both the sparsity level and the specific support, preferring smaller supports unless the data provides overwhelming evidence for density.

### 3. Comparison with Dense Halfspaces

**Statement.** If $\exists w^*$ with $\|w^*\|_0 \le s$ and $L_{\mathcal{D}}(w^*) \le \eta$, then any $n$ satisfying
$$n \gtrsim \frac{s \log(\tfrac{e d}{s}) + \log(\tfrac{1}{p_s}) + \log(\tfrac{1}{\delta})}{\epsilon^2}$$
ensures $L_{\mathcal{D}}(\hat{h}) \le \eta + \epsilon$ w.p. $\ge 1-\delta$. Dense halfspaces need $n \gtrsim \tfrac{d + \log(1/\delta)}{\epsilon^2}$. Sparse bound is smaller when $s \log(\tfrac{e d}{s}) \ll d$.

**Proof.** Plug $h=w^*$ into Part 2 bound and solve for $n$; compare with VC($d$) rate for dense halfspaces.

1. Part 2 with $k=s$ yields $L_{\mathcal{D}}(\hat{h}) \le \eta + C \sqrt{\tfrac{s \log(\tfrac{e d}{s}) + \log(1/p_s) + \log(1/\delta)}{n}}$.
2. Enforce the square-root term $\le \epsilon$ to obtain the stated $n$.
3. For dense halfspaces, $\operatorname{VCdim}=d$, giving rate $\sqrt{\tfrac{d + \log(1/\delta)}{n}}$.
4. Sparse advantage holds when $s \log(\tfrac{e d}{s})$ grows sublinearly vs. $d$. $\blacksquare$

**Comparison to the sample size obtained by learning over all homogeneous halfspaces in $\mathbb{R}^𝑑$.**
- Dense halfspaces need $n \gtrsim \tfrac{d + \log(1/\delta)}{\epsilon^2}$.
- The sparse bound needs $n \gtrsim \tfrac{s \log(\tfrac{e d}{s}) + \log(1/p_s) + \log(1/\delta)}{\epsilon^2}$.
- Sparse wins whenever $s \log(\tfrac{e d}{s}) \ll d)$, e.g., $(s = O(\log d))$ or $(s = d^\alpha)$ with $(\alpha<1)$.
- SRM can to adapt to unknown sparsity, provided the penalty properly encodes the search over supports

### 4. Validation as Model Selection

**Statement.** Using split $S_1,S_2$ with $|S_1|=|S_2|=n/2$, ERM $w_k$ on $S_1$ for each $k$, and $\hat{k}$ minimizing $L_{S_2}(w_k)$, we have w.p. $\ge 1-\delta$,
$$L_{\mathcal{D}}(w_{\hat{k}}) \le \inf_{1 \le k \le d,\, \|w\|_0 \le k} \Big\{ L_{\mathcal{D}}(w) + C \sqrt{\tfrac{k \log(\tfrac{e d}{k}) + \log(\tfrac{d}{\delta})}{n}} \Big\}.$$

**Proof.** Condition on $S_1$; finite-class validation on $S_2$ plus VC control of ERM on $S_1$.

1. Condition on $S_1$. The $d$ candidates $\{w_k\}$ form a finite class for $S_2$.
2. [Ref: Weighted finite-class bound] with uniform weights $1/d$ on $S_2$: w.p. $\ge 1-\delta$, for all $k$, $L_{\mathcal{D}}(w_k) \le L_{S_2}(w_k) + \sqrt{\tfrac{\log d + \log(1/\delta)}{n}}$.
3. Optimality of $\hat{k}$ on $S_2$ gives $L_{S_2}(w_{\hat{k}}) \le L_{S_2}(w_k)$ for all $k$.
4. VC generalization on $S_1$: w.p. $\ge 1-\delta$, $L_{\mathcal{D}}(w_k) \le L_{S_1}(w_k) + C' \sqrt{\tfrac{k \log(\tfrac{e d}{k}) + \log(1/\delta)}{n}}$, and ERM implies $L_{S_1}(w_k) \le L_{S_1}(w)$ for any $\|w\|_0 \le k$.
5. Combine (2)-(4) with $|S_1|=|S_2|=n/2$ and union-bound $\delta$'s; absorb constants to reach the displayed bound. Cost $k \log(\tfrac{e d}{k})$ fits $w_k$; $\log d$ selects $k$. $\blacksquare$

**Comparison to penalty-based SRM.**

- **Data Usage & Fitting:** **SRM** uses the full sample $S$ of size $n$ to fit every hypothesis $h \in \mathcal{H}_k$. **Validation** splits the data, using only $S_1$ (size $n/2$) to fit the candidates $\{w_k\}$.
- **Complexity Selection:** **SRM** chooses $k$ by minimizing the penalized empirical risk (training error + penalty) on the full sample. **Validation** chooses $k$ by minimizing the holdout risk specifically on the unseen set $S_2$.
- **Statistical Price:** **Validation** pays a "sample size tax"; because it uses $n/2$ for both fitting and selection, the variance in the bounds increases by a factor of $\sqrt{2}$. It also incurs a $\log d$ cost to account for the finite choice among $d$ models.
- **SRM Trade-off:** **SRM** maintains higher efficiency by using all $n$ points for both tasks, but it pays a "complexity tax" via the $\log(1/p_k)$ term to avoid overfitting the combined classes.
- **Practical Advantage:** **Validation** is highly robust because it avoids the need to manually specify priors ($p_k$) or tune penalty constants. However, this simplicity comes at the cost of statistical efficiency, which is most noticeable when $n$ is small.

### 5. Computation of Sparse Feasibility

**Statement.** Brute-force realizable learning enumerates all supports $I \subseteq [d]$, $|I|=k$, solving halfspace feasibility on each. Runtime $= \binom{d}{k}\, T_{\text{feas}}(k,n)$; polynomial in $d$ only when $k$ is $O(1)$ or $O(\log d)$.

**Proof.** Count supports; multiply by per-support LP/feasibility cost.

1. Enumerate $\binom{d}{k}$ supports; for each, solve feasibility in $\mathbb{R}^k$ with $n$ constraints via LP in $\operatorname{poly}(k,n)$.
2. Total time $\binom{d}{k}\operatorname{poly}(k,n) \approx (\tfrac{e d}{k})^k \operatorname{poly}(k,n)$.
3. Polynomial in $d$ only when $k$ is constant or logarithmic.
4. Highlights gap: sample complexity scales with $k \log(\tfrac{e d}{k})$, but computational cost can be exponential in $k$. $\blacksquare$

**Brute-force algorithm.** For each subset $I \subseteq [d]$ with $|I|=k$ (there are $\binom{d}{k}$ of them): restrict every $x_i$ to coordinates in $I$ and solve the linear feasibility problem:
$$\text{Find } w_I \in \mathbb{R}^k \text{ s.t. } y_i \langle w_I, x_i|_I \rangle > 0, \quad \forall i.$$
Finally, output any feasible $w_I$ after lifting back to $\mathbb{R}^d$ by padding zeros outside $I$.

**Runtime estimate.** Enumerating supports costs $\binom{d}{k}$. Feasibility for a fixed $I$ can be solved in time polynomial in $n$ and $k$ (e.g., via LP in $O(\text{poly}(n,k))$). Total runtime
$$T(d,k,n) = \binom{d}{k} \cdot \text{poly}(n,k).$$
This is polynomial in $d$ iff $k = O(1)$ or $k = O(\log d)$; once $k = \Theta(d)$ the binomial term is exponential in $d$.

**Distinction between sample complexity and computational complexity.**

Sample complexity depends on $k \log(\tfrac{e d}{k})$ (mild for moderate $k$), but brute-force enumeration is $\binom{d}{k} \cdot \operatorname{poly}(k,n)$, exponential in $k$; thus statistical feasibility does not imply computational tractability.

## Part B: PAC-Bayes for Thresholds

### 1. VC Dimension and Point-Posterior PAC-Bayes

**Statement.** For thresholds $\mathcal{H}_N$ on $\{1,\dots,N\}$, $\operatorname{VCdim}=1$. With uniform prior $P$ and point posterior $Q_S=\delta_{\hat{h}_t}$ from a consistent ERM,
$$KL(Q_S\parallel P)=\log(N+1), \quad L_{\mathcal{D}}(Q_S) \le L_S(Q_S)+\sqrt{\tfrac{\log(N+1)+\log(\tfrac{2n}{\delta})}{2(n-1)}}.$$

**Proof.** Show VC=1; compute KL for point mass under uniform prior; plug into [Ref: PAC-Bayes].

1. Two labeled points with opposite labels cannot both be realized by a single threshold unless ordered; only single point shatterable, so VC=1.
2. Uniform prior: $P(h_t)=1/(N+1)$; point posterior gives $KL=\log(N+1)$.
3. Realizable $S$ gives $L_S(Q_S)=0$; substitute into [Ref: PAC-Bayes].
4. The bound scales with $\log(N+1)$ because KL accounts for hypothesis count, unlike VC realizable bounds that depend solely on VC=1. $\blacksquare$

**Why the PAC-Bayes certificate scales with $log(𝑁 + 1)$ even though a VC-style realizable guarantee for thresholds should not scale with $log (𝑁)$.**

PAC-Bayes must pay for the description length of the selected hypothesis under the fixed prior. VC realizable bounds do not, because they rely on growth functions rather than information-theoretic weights. Hence the mismatch: even though thresholds are extremely simple, a naive uniform prior distributes mass across $N+1$ hypotheses and incurs $\log(N+1)$ complexity.

### 2. Version-Space Posterior

**Statement.** 

Let $a(S)=\max\{x_i: y_i=0\}$ (0 if none) and $b(S)=\min\{x_i: y_i=1\}$ ($N+1$ if none).

Then $V(S)=\{t: a(S)< t \le b(S)\}$ with $|V(S)|=b(S)-a(S)$.

For uniform prior $P$, uniform posterior $Q_V$ over $V(S)$ satisfies $KL(Q_V\parallel P)=\log(N+1)-\log|V(S)|$ and $L_S(Q_V)=0$.

**Proof.** Characterize consistency interval; compute KL of uniform-on-subset; note zero empirical loss.

1. Consistency requires threshold above largest negative and at/below smallest positive, giving the interval form.
2. $Q_V(h_t)=1/|V(S)|$, $P(h_t)=1/(N+1)$; KL simplifies to $\log(N+1)-\log|V(S)|$.
3. All $h_t$ in $V(S)$ classify $S$ correctly, so $L_S(Q_V)=0$.
4. Larger $|V(S)|$ lowers KL versus a point posterior, tightening the PAC-Bayes certificate. $\blacksquare$

**Why $𝑄𝑉$ can give a better PAC-Bayes bound than a point posterior when $|𝑉(𝑆)|$ is large.**

When $|V(S)|$ is large, spreading the posterior uniformly over $V(S)$ reduces the KL term by $\log|V(S)|$ while keeping $L_S(Q_V)=0$, tightening the PAC-Bayes bound relative to a point posterior.

### 3. Spreading Helps KL, May Hurt True Risk

**Statement.** For realizable $\mathcal{D}$ with true threshold $h_{\tau}$ (uniform $X$), any nonempty $W \subseteq \{1,\dots,N+1\}$ and uniform posterior $Q_W$ satisfy
$$L_{\mathcal{D}}(Q_W)= \tfrac{1}{N|W|} \sum_{t \in W} |t-\tau|. $$
There exist realizable $S$ with large $|V(S)|$ such that $KL(Q_{V(S)}\parallel P) < KL(\delta_{h_{\tau}}\parallel P)$ yet $L_{\mathcal{D}}(Q_{V(S)}) > L_{\mathcal{D}}(\delta_{h_{\tau}})$.

**Proof.** Compute disagreement mass; exhibit wide interval around $\tau$.

1. Disagreement of $h_t$ vs $h_{\tau}$ occurs on indices between $t$ and $\tau$, fraction $|t-\tau|/N$; averaging gives displayed risk.
2. Example: $N=20$, $\tau=10$, and realizable $S$ leaving $V(S)=\{6,\dots,14\}$ (size 9). Then $KL(Q_{V(S)}\parallel P)=\log 21-\log 9 < \log 21 = KL(\delta_{h_{10}}\parallel P)$.
3. Risk: $\tfrac{1}{20\cdot 9}\sum_{t=6}^{14} |t-10| = \tfrac{53}{180} > 0$, while $L_{\mathcal{D}}(\delta_{h_{10}})=0$.
4. Lower KL can coexist with higher true risk; PAC-Bayes bounds certify via KL, not monotonic in true risk across posteriors. $\blacksquare$

**Why there is no contradiction with PAC-Bayes.**

A posterior with lower $KL$ can still have higher true risk if it spreads mass over suboptimal thresholds; PAC-Bayes bounds are driven by $KL$, not by monotonicity in true risk, so the example does not contradict PAC-Bayes. PAC-Bayes upper-bounds $L_{\mathcal{D}}(Q_W)$ by empirical loss plus a complexity term. Here $L_S(Q_{V(S)})=0$, and the complexity term remains nontrivial (though smaller than the point posterior). The theorem never promised that minimizing KL would minimize true risk—only that both terms together bound the risk.

### 4. Every Fixed Prior Can Be Attacked

**Statement.** For any prior $P$ on $\mathcal{H}_N$ with $N \ge 3$, some $\tau \in \{2,\dots,N\}$ has $P(h_{\tau}) \le 1/(N-1)$. For distribution $\mathcal{D}_{\tau}$ with mass $1/2$ on $(\tau-1,0)$ and $1/2$ on $(\tau,1)$, with probability $\ge 1-2^{1-n}$ over $S \sim \mathcal{D}_{\tau}^n$, both support points appear, so $V(S)=\{\tau\}$ and any zero-empirical-error posterior equals $\delta_{h_{\tau}}$, yielding $KL(Q\parallel P) \ge \log(N-1)$.

**Proof.** Pigeonhole on prior mass; bound missing-support probability; uniqueness of consistent threshold fixes posterior.

1. Internal thresholds sum to $\le 1$; pigeonhole gives some $\tau$ with $P(h_{\tau}) \le 1/(N-1)$.
2. Each support point missing with prob. $(1/2)^n$; union bound gives both seen with prob. $\ge 1-2^{1-n}$.
3. When both seen, consistency forces $V(S)=\{\tau\}$; any $Q$ with $L_S(Q)=0$ is $\delta_{h_{\tau}}$.
4. Then $KL(Q\parallel P)=\log\tfrac{1}{P(h_{\tau})} \ge \log(N-1)$.
5. This limits zero-empirical-error PAC-Bayes certificates with fixed prior, not sample complexity of thresholds (VC=1). $\blacksquare$

**Why it is a limitation of zero-empirical-error PAC-Bayes certificates with a fixed prior, rather than a lower bound on the sample complexity of learning thresholds?**

The lower bound shows that with a fixed prior and zero empirical error, $KL$ can be forced to $\Omega(\log N)$ even though thresholds have $VC=1$. It is a limitation on zero-error PAC-Bayes certificates with a fixed prior, not a lower bound on the sample complexity of learning thresholds (which remains constant-VC learnable). This is not a lower bound on sample complexity for learning thresholds; efficient VC algorithms achieve perfect generalization with $O(\frac{\log(1/\delta)}{\epsilon})$ samples independent of $N$. Rather, it shows a limitation of PAC-Bayes certificates that insist on zero empirical error with a fixed prior: an adversary can force large KL, hence loose bounds. Allowing nonzero empirical loss or data-dependent priors can evade this limitation.
