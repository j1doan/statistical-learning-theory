## Part A: Unions of Two Intervals on the Line

### 1. Restriction patterns.

A labeling $(y_1, \dots, y_n) \in \{0, 1\}^n$ of ordered points $x_1 < x_2 < \dots < x_n$ is realizable by $\mathcal{H}_2$ if and only if it contains at most **two** contiguous blocks of 1s.

**Proof:**
A hypothesis $h \in \mathcal{H}_2$ is defined by $h(x) = \mathbf{1}[x \in I_1 \cup I_2]$, where $I_1, I_2$ are intervals in $\mathbb{R}$. 
1. **Sufficiency:** If a labeling has at most two blocks of 1s, say $\{x_i, \dots, x_j\}$ and $\{x_k, \dots, x_l\}$ (where $j < k-1$), we can choose $I_1 = [x_i, x_j]$ and $I_2 = [x_k, x_l]$. Points within these intervals are labeled 1, and points outside (including those between $x_j$ and $x_k$) are labeled 0. If there is only one block, we can set $I_2 = \emptyset$. If there are zero blocks (all 0s), we set $I_1 = I_2 = \emptyset$.
2. **Necessity:** An interval $I \subseteq \mathbb{R}$ can capture at most one contiguous block of 1s from an ordered set of points. Any points $x_a < x_b < x_c$ such that $x_a, x_c \in I$ implies $x_b \in I$ by the definition of an interval. Thus, the union of two intervals $I_1 \cup I_2$ can capture at most two such contiguous blocks. 

The structural property is the **number of 0-to-1 transitions (starts of blocks)**, which must be $\le 2$.

### 2. Exact growth function.

The growth function $\Gamma_{\mathcal{H}_2}(n)$ is the number of distinct labelings of $n$ points realizable by $\mathcal{H}_2$. From our characterization, these are labelings with 0, 1, or 2 blocks of 1s.

**Derivation:**
A labeling is determined by the start and end indices of the blocks. Let the points be $x_1, \dots, x_n$. We can imagine $n+1$ possible "slots" for boundaries (including before $x_1$ and after $x_n$). However, it is easier to count by the number of blocks:
1. **0 blocks:** 1 labeling (all 0s).
2. **1 block:** A block is defined by choosing 2 boundaries out of $n+1$ possible positions (the gaps between points and the ends). There are $\binom{n+1}{2}$ such labelings.
3. **2 blocks:** Two disjoint blocks are defined by choosing 4 boundaries out of $n+1$ positions. There are $\binom{n+1}{4}$ such labelings.

Thus, the exact growth function is:
$$\Gamma_{\mathcal{H}_2}(n) = \binom{n+1}{4} + \binom{n+1}{2} + 1$$

**Proof by Induction/Combinatorics:**
Any labeling with at most 2 blocks can be uniquely represented by choosing $2k$ distinct boundary positions from the $n+1$ available spaces between/around points, where $k \in \{0, 1, 2\}$ is the number of blocks. 
- $k=0$: $\binom{n+1}{0} = 1$
- $k=1$: $\binom{n+1}{2}$
- $k=2$: $\binom{n+1}{4}$
Summing these gives the total number of realizable labelings.

### 3. VC dimension and mistake bound.

**VC Dimension:**
The VC dimension is the largest $d$ such that there exists a set $S$ of size $d$ that is shattered by $\mathcal{H}_2$ (equivalently, $|\mathcal{H}_{2\mid S}|=2^d$).
- For $n=4$: $\Gamma_{\mathcal{H}_2}(4) = \binom{5}{4} + \binom{5}{2} + 1 = 16 = 2^4$, so some 4-point set is shattered.
- For $n=5$: $\Gamma_{\mathcal{H}_2}(5) = \binom{6}{4} + \binom{6}{2} + 1 = 31 < 2^5$, so no 5-point set is shattered.
Thus, **$VCdim(\mathcal{H}_2) = 4$**.

**Halving Mistake Bound:**
In the realizable online transductive setting with a pool $X$ of $n$ points, the Halving algorithm makes at most $\log_2 |\mathcal{H}_{2\mid X}|$ mistakes. Hence, in the worst case over pools,
$$M \le \log_2 |\mathcal{H}_{2\mid X}| \le \log_2 \Gamma_{\mathcal{H}_2}(n) = \log_2\left(\binom{n+1}{4} + \binom{n+1}{2} + 1\right).$$

**Comparison with Sauer-Shelah:**
The Sauer-Shelah bound states $\Gamma_{\mathcal{H}}(n) \le \sum_{i=0}^{d} \binom{n}{i}$. For $d=4$:
$$SS(n, 4) = \binom{n}{4} + \binom{n}{3} + \binom{n}{2} + \binom{n}{1} + \binom{n}{0}$$
Our formula: $\Gamma_{\mathcal{H}_2}(n) = \binom{n+1}{4} + \binom{n+1}{2} + 1$.
Using the identity $\binom{n+1}{k} = \binom{n}{k} + \binom{n}{k-1}$:
$\Gamma_{\mathcal{H}_2}(n) = [\binom{n}{4} + \binom{n}{3}] + [\binom{n}{2} + \binom{n}{1}] + \binom{n}{0} = \sum_{i=0}^{4} \binom{n}{i}$.

The formulas are **equal**. This tells us that the Sauer-Shelah bound is **tight** for this hypothesis class. This happens when the class is a "maximum class" whose shatter function meets the Sauer--Shelah upper bound.

### 4. Tightness of Sauer-Shelah.

For arbitrary $n \ge d \ge 0$, let $\mathcal{X} = \{1, \dots, n\}$. Define $\mathcal{H}$ as the class of all subsets of $\mathcal{X}$ with size at most $d$:
$$\mathcal{H} = \{ A \subseteq \mathcal{X} : |A| \le d \}$$

**Verification:**
1. **VC Dimension:** Any subset of size $d$ is shattered because $\mathcal{H}$ contains all its subsets. No subset of size $d+1$ is shattered because the labeling $(1, 1, \dots, 1)$ (all $d+1$ points) is not in $\mathcal{H}$. Thus $VCdim(\mathcal{H}) = d$.
2. **Growth Function:** The number of realizable labelings on $n$ points is simply the number of subsets in $\mathcal{H}$:
$$\Gamma_{\mathcal{H}}(n) = \sum_{k=0}^{d} \binom{n}{k}$$

**Conclusion:**
This example shows that the Sauer-Shelah bound is the **best possible upper bound** that can be stated solely in terms of $n$ and $d$, as there exist natural hypothesis classes that achieve the bound exactly for all $n, d$.

### 5. AI proof audit.

**Critique of the AI's argument:**
1. **Incomplete Boundary Definition:** Switches occur in the **gaps** between points or at the ends, not at the points themselves. There are $n+1$ potential boundary locations (before $x_1$, between consecutive points, after $x_n$).
2. **Pairing Constraint:** Switches must come in **pairs** (start/end of a block of 1s). Counting "up to 4 switch locations" among $n$ positions ignores that only even numbers of boundaries are valid and that boundaries live in the $n+1$ gaps.
3. **Missing Surjectivity Argument:** The AI does not show that every valid boundary choice corresponds to an $\mathcal{H}_2$ hypothesis and that every $\mathcal{H}_2$ hypothesis is counted exactly once.

**Corrected Statement:**
"A labeling in $\mathcal{H}_2$ is characterized by having at most two contiguous blocks of 1s. Such a labeling is uniquely determined by choosing an even number of transition boundaries $2k$ (where $k \in \{0, 1, 2\}$) from the $n+1$ gaps. This yields
$$\Gamma_{\mathcal{H}_2}(n) = \binom{n+1}{4} + \binom{n+1}{2} + 1,$$
and hence $\Gamma_{\mathcal{H}_2}(n)=\sum_{j=0}^4 \binom{n}{j}$ by Pascal's identity. Since $\Gamma(4) = 2^4$ and $\Gamma(5) < 2^5$, the VC dimension is exactly 4."

## Part B: Quadratic Threshold Functions on the Line

### 1. A general upper-bound trick.

**Theorem:** If $h(x) = \mathbf{1}[\langle w, \phi(x) \rangle \ge 0]$ for $\phi: \mathcal{X} \to \mathbb{R}^D$, then $VCdim(\mathcal{H}) \le D$.

**Proof:**
Consider the class of homogeneous halfspaces in $\mathbb{R}^D$:
$$\mathcal{H}_{half} = \{ z \mapsto \mathbf{1}[\langle w, z \rangle \ge 0] : w \in \mathbb{R}^D \}$$
We are given that $VCdim(\mathcal{H}_{half}) = D$. 
Any labeling of points $x_1, \dots, x_n \in \mathcal{X}$ produced by $h \in \mathcal{H}$ is equivalent to a labeling of points $z_1, \dots, z_n \in \mathbb{R}^D$ (where $z_i = \phi(x_i)$) produced by some $h' \in \mathcal{H}_{half}$. 
Therefore, the number of labelings $\mathcal{H}$ can produce on any $n$ points in $\mathcal{X}$ is at most the number of labelings $\mathcal{H}_{half}$ can produce on $n$ points in $\mathbb{R}^D$. 
$$\Gamma_{\mathcal{H}}(n) \le \Gamma_{\mathcal{H}_{half}}(n)$$
Since $\mathcal{H}_{half}$ cannot shatter any $D+1$ points, $\mathcal{H}$ also cannot shatter any $D+1$ points. Thus, $VCdim(\mathcal{H}) \le VCdim(\mathcal{H}_{half}) = D$.

### 2. Apply the trick to quadratic thresholds.

**Transformation:**
A hypothesis in $\mathcal{H}_{quad}$ is $h(x) = \mathbf{1}[ax^2 + bx + c \ge 0]$. 
Define $\phi: \mathbb{R} \to \mathbb{R}^3$ as:
$$\phi(x) = (x^2, x, 1)$$
And let $w = (a, b, c) \in \mathbb{R}^3$. Then:
$$ax^2 + bx + c = \langle w, \phi(x) \rangle$$
So $h(x) = \mathbf{1}[\langle w, \phi(x) \rangle \ge 0]$.

**Upper Bound:**
By the general upper-bound trick with $D=3$, we immediately have:
$$VCdim(\mathcal{H}_{quad}) \le 3$$

### 3. Exact restriction pattern.

A quadratic polynomial $p(x) = ax^2 + bx + c$ can have at most 2 real roots. This determines the shape of the region $p(x) \ge 0$:
1. **$a > 0$:** The region is $(-\infty, r_1] \cup [r_2, \infty)$ (outside the roots) or $\mathbb{R}$ (no roots/one root).
2. **$a < 0$:** The region is $[r_1, r_2]$ (between the roots) or $\emptyset$ (no roots).
3. **$a = 0$:** The polynomial is linear, $bx + c \ge 0$, giving a ray $[r, \infty)$, $(-\infty, r]$, $\mathbb{R}$, or $\emptyset$.

**Characterization:**
A labeling of $x_1 < \dots < x_n$ is realizable by $\mathcal{H}_{quad}$ if and only if it is either
- **At most one block of 1s** (possibly empty, so all-0 and all-1 are allowed), or
- **Two blocks of 1s that touch both ends** (equivalently, the complement of a single block of 1s in the middle).

Equivalently, along the ordered sample the labels have at most two transitions, and if there are two transitions then the labeling can be either $0\to1\to0$ (concave down) or $1\to0\to1$ (concave up), matching the sign of $a$.

### 4. Exact growth and exact VC dimension.

**Growth Function:**
Count labelings by the number of transitions in the ordered sample. There are $n-1$ gaps between consecutive points.
- **0 transitions:** 2 labelings (all 0s or all 1s).
- **1 transition:** choose one gap and choose the starting label, giving $2\binom{n-1}{1}=2(n-1)$ labelings.
- **2 transitions:** choose two gaps and choose whether the pattern is $0\to1\to0$ or $1\to0\to1$, giving $2\binom{n-1}{2}$ labelings.

Therefore
$$\Gamma_{\mathcal{H}_{quad}}(n) = 2 + 2\binom{n-1}{1} + 2\binom{n-1}{2} = n^2 - n + 2.$$

**VC Dimension:**
- $n=3$: $\Gamma(3) = 3^2 - 3 + 2 = 8 = 2^3$. Shattered.
- $n=4$: $\Gamma(4) = 4^2 - 4 + 2 = 14 < 2^4 = 16$. Not shattered.
Thus, **$VCdim(\mathcal{H}_{quad}) = 3$**.

**Comparison with Sauer-Shelah:**
For $d=3$:
$SS(n, 3) = \binom{n}{3} + \binom{n}{2} + \binom{n}{1} + \binom{n}{0} = \frac{n(n-1)(n-2)}{6} + \frac{n(n-1)}{2} + n + 1$.
Our $\Gamma(n) = n^2 - n + 2$ is much smaller than the cubic Sauer-Shelah bound for large $n$. 

**Conclusion:**
The VC-based bound is **not tight** here. This shows that VC dimension is a coarse summary; two classes can have the same VC dimension but very different growth rates. $\mathcal{H}_{quad}$ is "thinner" than the maximum class of dimension 3.

### 5. AI proof audit.

**Critique of the AI's argument:**
1. **Correct Observation, Wrong Counting:** The AI notes there are at most two transitions, but it counts only the number of ways to select transition gaps and ignores the starting label.
2. **Missing Symmetry:** For any set of transition gaps, there are **two** possible labelings (e.g., $010$ and $101$), corresponding to $a<0$ vs. $a>0$. Thus the count must be multiplied by 2.
3. **VC Dimension Mismatch:** Plugging $n=3$ into the AI's formula gives 4 labelings, which would force $VCdim<3$, contradicting its stated conclusion.

**Corrected Theorem:**
"A quadratic polynomial $p(x)$ has at most two real roots, implying that the labeling of an ordered sample can have at most two transitions. For any choice of $k \in \{1, 2\}$ transition gaps among the $n-1$ available, there are exactly 2 possible labelings (starting with 0 or 1). Including the 2 constant labelings, we have $\Gamma_{\mathcal{H}_{quad}}(n) = 2\binom{n-1}{2} + 2\binom{n-1}{1} + 2 = n^2 - n + 2$. This confirms $VCdim(\mathcal{H}_{quad}) = 3$ as $\Gamma(3)=8$ and $\Gamma(4)=14$."
